```python
from selenium.webdriver import Chrome, Firefox
import time
```

```python
from selenium import webdriver #  ~
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized"); #   
browser = webdriver.Chrome('chromedriver', options=options)
```

###  

```python
browser.get('https://www.data.go.kr/') #  
browser.implicitly_wait(5) #  5 (  )
```

```python
browser.find_element_by_xpath('//*[@id="header"]/div/div/div/div[2]/div/a[1]').click() #  ( xpath)
browser.implicitly_wait(5)
```

```python
browser.find_element_by_xpath('//*[@id="mberId"]').send_keys('winston1214') # id  ("USER_ID")
```
