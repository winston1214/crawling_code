# 크롬 띄우기

```python
from selenium.webdriver import Chrome, Firefox
import time
```

```python
from selenium import webdriver # 닫지 말아요~
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized"); # 시작할때 최대창을 띄워라
browser = webdriver.Chrome('chromedriver', options=options)
```

### 페이지 접속

```python
browser.get('https://www.data.go.kr/') # 사이트 접속
browser.implicitly_wait(5) # 묵시적으로 5초동안 대기해라(쪼매 줄여도 됨)
```

```python
browser.find_element_by_xpath('//*[@id="header"]/div/div/div/div[2]/div/a[1]').click() # 로그인을 클릭해라(로그인 xpath긁어옴)
browser.implicitly_wait(5)
```

```python
browser.find_element_by_xpath('//*[@id="mberId"]').send_keys('winston1214') # id 넣고 ("USER_ID쪽에")
```

```python
browser.find_element_by_xpath('//*[@id="pswrd"]').send_keys('winston2605!') # 비밀번호 넣고('PASSWORD쪽')
```

```python
browser.find_element_by_xpath('//*[@id="loginVo"]/div[2]/div[2]/div[2]/div/div[1]/button').click() # 로그인 버튼 클릭
browser.implicitly_wait(5) 
```

```python
browser.find_element_by_xpath('//*[@id="M000400_pc"]/a').click() # 정보공유 클릭
```

```python
browser.find_element_by_xpath('//*[@id="M000402_pc"]/a').click() # 자료실 클릭
```

```python
trs = browser.find_elements_by_xpath('//*[@id="searchVO"]/div[5]/table/tbody/tr')
# //*[@id="searchVO"]/div[5]/table/tbody/tr[1]/td[2]/a
for tr in trs:
    print(tr.find_element_by_xpath('td[2]').text)    
```

```python
browser.find_element_by_xpath('//*[@id="searchVO"]/div[5]/table/tbody/tr[1]/td[2]/a').click()
browser.implicitly_wait(3) # 처음으로 들어가라
```

```python
browser.find_element_by_xpath('//*[@id="recsroomDetail"]/div[2]/div[4]/div/a').click() # 첨부파일 클릭
time.sleep(10) # 실행보류
```

```python
browser.quit() # 반드시 끝낼때 quit해라
```
